## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
Coords = boston[['LATITUDE', 'LONGITUDE']]
scaler = StandardScaler().fit(Coords)
CoordsScaled = scaler.transform(Coords)
```

```python
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')

evaluate_clusters(CoordsScaled, 10)
```

Looking at the graph, the optimal number of clusters appears to be about 4, as you gain a great improvement in error by adding more clusters up to 4, where the decrease in error slows down significantly.

```python
k = 4
kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
assignment = kmeans.fit_predict(CoordsScaled)    
cmap = sns.hls_palette(n_colors=k, h=0.35, l=0.4, s=0.9)

print len(CoordsScaled[:,0])

for i in range(0,k):
    Lat = []
    Long = []
    for j in range(0,len(assignment)):
        if(assignment[j] == i):
            Lat.append(CoordsScaled[j, 1])
            Long.append(CoordsScaled[j, 0])
    plt.scatter(Lat, Long, c=cmap[i])

```

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering.
```python
on_time = pd.get_dummies(boston['OnTime_Status'])
#status = pd.get_dummies(boston['CASE_STATUS'])
#department = pd.get_dummies(boston['Department'])
d = {'LATITUDE' : CoordsScaled[:,0], 'LONGITUDE': CoordsScaled[:,1]}
CoordsDF = pd.DataFrame(d)
concatenated = pd.concat([CoordsDF, on_time], axis = 1 )


evaluate_clusters(concatenated, 12)
```

Here I filter the dataset by taking the features for the on time status of the record, whether the record is open or closed, as well as the geolocation. To use the categorical features OnTime_status and Case_Status, I use the pandas get_dummies function, which converts a categorical feature to muliple columns, where each column represents one of the possible options for the category, where for a record that had a value 'a' for the feature will have a 1 in the 'a' column, and 0 in the other created columns. This allows us to treat the categorical data as numeric. We can then concatenate all the data along the axis = 1, and evaluate the clusters in the same way as above.

Examining the resulting graph, an appropriate number of clusters here would be 8.

```python
k = 8

kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
assignment = kmeans.fit_predict(concatenated)    
cmap = sns.hls_palette(n_colors=k*2, h=0.35, l=0.4, s=0.9)
for i in range(0,k):
    LatOn = []
    LongOn = []
    LatOver = []
    LongOver= []
    for j in range(0,len(assignment)):
        if(assignment[j] == i):
            if(on_time['ONTIME'][j] == 1):
                LatOn.append(CoordsScaled[j, 1])
                LongOn.append(CoordsScaled[j, 0])
            else:
                LatOver.append(CoordsScaled[j, 1])
                LongOver.append(CoordsScaled[j, 0])
    plt.scatter(LatOn, LongOn, color = 'blue')
    plt.scatter(LatOver, LongOver, color = 'red')
    plt.axis([-3, 3, -3, 3])
    plt.show()
```
